```python
import pandas as pd
from dash import Dash,dcc,html
import plotly.express as px
```

```python
df = pd.read_csv(r"C:/Users/user/Downloads/archive/train.csv",delimiter=';')
```

```python
df.head(5)
```

```python
app = Dash(__name__)
```

```python
marital_df = df.groupby(['marital'])['marital'].count().reset_index(name="count")
marital_df.columns=['marital','count']
```

```python
fig=px.bar(marital_df,x='marital',y='count',title='marital profile')
fig.show()
```

```python
jobs_df = df.groupby(['job'])['job'].count().reset_index(name='count')
```

```python
jobs_df.head(5)
```

```python
fig2=px.pie(df,values='balance',names='job',title='Job Profile & Balances')
fig2.show()
```

```python
app.layout = html.Div([
    html.Div([
        html.Div([
            html.H3('Data 1'),
            dcc.Graph(id='g1', figure=fig)
        ], className="six columns"),

        html.Div([
            html.H3('Data 2'),
            dcc.Graph(id='g2', figure=fig2 )
        ], className="six columns"),
    ], className="row")
])

app.css.append_css({
    'external_url': 'https://codepen.io/chriddyp/pen/bWLwgP.css'
})

if __name__ == '__main__':
    app.run_server(debug=True)
```
