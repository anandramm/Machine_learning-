```
import pandas as pd 
import numpy as np
```

```
data=pd.read_csv('/content/airline_passengers.csv',index_col='Month',parse_dates=True)
```

```
data.count()
```

```
data.dropna()
```

```
data.index
```

```
data.index.freq='MS'
```

```
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```
span=12
alpha=2/(1+span)
```

```
data['EWMA12']=data['Thousands of Passengers'].ewm(alpha=alpha,adjust=False).mean()
```

```
data.head()
```

```
model=SimpleExpSmoothing(data['Thousands of Passengers'])
```

```
fitted_mode=model.fit(smoothing_level=alpha,optimized=False)
```

```
data['SES12']=fitted_mode.fittedvalues.shift(-1)
```

```
data.head()
```

```
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```
model_1=ExponentialSmoothing(data['Thousands of Passengers'],trend='add')
```

```
fm=model_1.fit()
```

```
data['DES12']=fm.fittedvalues.shift(-1)
```

```
data.head()
```

```
data['TES_MUL_12']=ExponentialSmoothing(data['Thousands of Passengers'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
```

```
data.head()
```

```
data.plot()
```

```
import pandas as pd
import numpy as np
```

```
dt=pd.read_csv('/content/EnergyProduction.csv',index_col='DATE',parse_dates=True)
```

```
dt.isna().sum()
```

```
dt.asfreq(freq='MS')
```

```
dt.index
```

```
dt.plot(figsize=(20,7))
```

```
dt['SMA 12']=dt['EnergyIndex'].rolling(window=12).mean()
```

```
dt.plot(figsize=(20,7))
```

```
dt['EWM 12']=dt['EnergyIndex'].ewm(alpha=(2/(1+12)),adjust=False).mean()
```

```
dt.plot(figsize=(20,7))
```

```
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```
model_time=ExponentialSmoothing(dt['EnergyIndex'],trend='add',seasonal='add',seasonal_periods=12)
```

```
final=model_time.fit()
```

```
dt['TES 12']=final.fittedvalues
```

```
dt.plot(figsize=(20,6))
```
